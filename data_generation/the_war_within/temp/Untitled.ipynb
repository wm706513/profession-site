```python
def get_specializations(profession):
    prof_data = dict()
    
    stats = ['skill', 'multicraft', 'resourcefulness', 'ingenuity']
    for stat in stats:
        prof_data[stat] = dict()
        with open(DATA_DIR+profession.lower()+'_specializations_'+stat+'.csv', 'r') as file:
            data = file.readlines()[1:]
            for line in data:
                line = line.strip().split(',')
                node = line[0]
                maxPoints = int(line[1])
                gainPerPoint = int(line[2])
                gains = line[3:][:maxPoints//5+1]
                gains = [int(x) for x in gains]
                    
                prof_data[stat][node] = dict(zip(np.arange(start=0, stop=maxPoints+1, step=5), gains))
                prof_data[stat][node]['scaling'] = gainPerPoint
                
    return prof_data
```

```python
def get_items(profession):
    prof_items = dict()
    
    with open(DATA_DIR+profession.lower()+'_specializations_items.csv', 'r') as file:
        data = file.readlines()[1:]
        
        for line in data:
            line = line.strip().split(',')
            itemID = int(line[0])
            itemName = line[1]
            
            lastIndex = len(np.where(np.array(line)!='')[0])
            nodes = line[2:lastIndex]
            
            prof_items[itemID] = dict()
            prof_items[itemID]['itemName'] = itemName
            prof_items[itemID]['nodes'] = nodes
            
    return prof_items
```

```python
get_items('alchemy')
```

```python
def calculate_stat(stat, profession, itemID):
    nodes = get_items(profession)[itemID]['nodes']
    baseStat = 0
    level = 100
    equipment = 34
    points = get_specializations(profession)[stat]
    knowledge = alchemy1
    finalStat = baseStat + level + equipment
    
    for node in nodes:
        appliedKnowledge = knowledge.loc[knowledge['node']==node, 'points'].iloc[0]
        
        if appliedKnowledge == -1:
            continue
        
        finalStat += points[node]['scaling']*appliedKnowledge
        for level in np.arange(start=0, stop=appliedKnowledge+1, step=5):
            finalStat += points[node][level]
            
    return finalStat
```
